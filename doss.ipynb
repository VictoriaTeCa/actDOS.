```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('Inmuebles2.csv') 
data
```

```python
#Eliminar primera fila o filas innecesarias
data1= pd.read_csv('Inmuebles2.csv', skiprows=1, low_memory=False)
data1
```

```python
#Eliminar columnas innecesarias
data2= data1.drop(['0','Referencia'], axis=1)
data2
```

```python
#Identificar valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data3=data2.copy()
data3
```

```python
#Sustituir valores nulos con promedio o media
data3["Superficie"]=data3["Superficie"].fillna(round(data2["Superficie"].mean(),1))
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana
data3["Precio Venta"]=data3["Precio Venta"].fillna(round(data2["Precio Venta"].median(),1))
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Eliminamos filas con nulos
data4= data3.dropna(0)
data4
```

```python
#Corroboramos valores nulos
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#sustituir valores nulos por un valor en concreto
#filtro por columnas
#
data3["Fecha Alta"]=data3["Fecha Alta"].fillna(10)
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3["Fecha Venta"]=data3["Fecha Venta"].fillna("2004-06-19 00:00:00")
data3
```

```python
data3["Tipo"]=data3["Tipo"].fillna(method="bfill")
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3["Operación"]=data3["Operación"].fillna(method="bfill")
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3["Provincia"]=data3["Provincia"].fillna(method="bfill")
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia hacia adelante forward fill ("ffill")
data3["Vendedor"] = data3 ["Vendedor"].fillna(method="ffill")
data3
```
